## Reading from the file and storing it in an array

```python
import nltk
import string
import re
from nltk.corpus import stopwords
from collections import Counter

def read_file(file_name):
    f = open(file_name,"r")
    docs = f.readlines()
    return docs
docs = read_file("documents.txt")
docs
```

## Tokenization of the words and removal of any non-alphanumeric character

```python
def tokenize_to_array(array):
    tokenized_array = []
    for i in docs:
        if i.strip():
            i = i.replace('\\n', '')
            #i = i.replace('',"")
            i = re.sub("[^a-zA-Z0-9]+", " ", i) #removing all symbols except alphanumeric characters
            tokenized_array.append(i)
    return tokenized_array

        

tokenized_array = tokenize_to_array(docs)
tokenized_array
```

## Separating the doc ids from the text 

```python
import re

def text_compile(array):
    
    normalized_text = []
    #this step extracts only the text that is between <DOC> and </DOC>
    lines = re.compile('DOC (.*?) DOC ', re.DOTALL |  re.IGNORECASE ).findall(str(array))
    for i in lines:
        i = i.rstrip()
        normalized_text.append(i.lower())
    return normalized_text

text_compiled = text_compile(tokenized_array)
text_compiled
```

## Converting the array to a dictionary with the doc ids as the keys and corresponding text as the values

```python
def convert_to_dicionary(text):
    count = 1
    dictionary = {}
    for i in text:
    #print(i[6:])
        dictionary['doc'+str(count)] = i[6:]
        count += 1
    return dictionary
dictionary = convert_to_dicionary(text_compiled)
    
dictionary
```

## Stemming and stopwords removal

```python
from nltk.corpus import stopwords
from collections import defaultdict
from collections import Counter

    
def stemming_and_stopwords_removal(words):
    stop_words = ['the', 'is', 'at', 'of', 'on', 'and', 'a']
    #stemmer = nltk.stem.snowball.SnowballStemmer("english")
    stemmer = nltk.stem.porter.PorterStemmer()
    final_arr = []
    for v in nltk.tokenize.word_tokenize(words):
        v = re.sub("[^a-zA-Z0-9]+", " ", v)
        v = v.lower()
        v = v.replace(",","")
        v = v.replace('',"")
        v = v.strip()
        stemmed = stemmer.stem(v)
        if stemmed not in stop_words and stemmed and stemmed!= 'dict valu':
            final_arr.append(stemmed) 

    return final_arr

final_arr = stemming_and_stopwords_removal(str(dictionary.values()))
final_arr

    
```

## Creating the final inverted index dictionary with the terms as the keys and the document list as the values

```python

def convert_to_inverted_index_dictionary(final):
    my_dict = defaultdict(list)
    for k,v in dictionary.items():
        for i in final:
            if i in v:
                my_dict[i].append(k)
    return my_dict
#stemming_and_stopwords_removal(dictionary,stemmer,stop_words)

my_dict = convert_to_inverted_index_dictionary(set(final_arr))

my_dict
```

## Writing the inverted index to a csv file

```python
import csv
csv_columns = ['Keyword','Document Frequency','Posting List']
with open('inverted_index.csv','w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(csv_columns)
    for k,v in sorted(my_dict.items()):
        writer.writerow([k,len(v),str(v).strip('[]')])
        #print(k,len(v),v)
    
```

## Performing the intersection on the queries

```python
def Intersect(p1,p2):
    answer = set()
    i = 0
    j = 0
    if p1 not in dictt or p2 not in dictt:
        return "Query Not found"
    dictt[p1] = sorted(dictt[p1])
    dictt[p2] = sorted(dictt[p2])
    while dictt[p1][i] is not None and dictt[p2][j] is not None:
        if dictt[p1][i] == dictt[p2][j]:
            answer.add(dictt[p1][i])
            i += 1
            j += 1
        elif dictt[p1][i]< dictt[p2][j]:
            i += 1
        else:
            j += 1
        if i == len(dictt[p1]) or j == len(dictt[p2]):
            break

    return answer
```
